```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
 movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_actors(obj):
    A = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter!=3:
            A.append(i['name'])
            counter+=1
        else:break
    return A
```

```python
convert_actors(movies['cast'][0])
```

```python
movies['cast'] = movies['cast'].apply(convert_actors)
```

```python
movies.head()
```

```python
ast.literal_eval(movies['crew'][0])
```

```python
def convert_crew(obj):
    C = []
    for i in ast.literal_eval(obj):
        if(i['job']=='Director' or i['job']=='Writer' or i['job']=='Producer'):
            C.append(i['name'])
    return C
```

```python
movies['crew'] = movies['crew'].apply(convert_crew)
```

```python
movies.head()
```

```python
movies['overview'][0].lower()
```

```python
from nltk.corpus import stopwords
import string
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def pre_process(obj):
    obj = obj.lower()
    obj = nltk.word_tokenize(obj)
    y = []
    for i in obj:
        if i.isalnum():
            y.append(i)
            
    obj = y[:]
    y.clear()
    
    for i in obj:
        if i not in stopwords.words('english') and string.punctuation:
            y.append(ps.stem(i))
            
    return " ".join(y)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_movies = movies[['movie_id', 'title', 'tags']]
```

```python
new_movies
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : " ".join(x))
```

```python
new_movies
```

```python
new_movies['tags'] = new_movies['tags'].apply(lambda x : x.lower())
```

```python
new_movies.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer(max_features = 5000)
```

```python
vectors = vec.fit_transform(new_movies['tags']).toarray()
```

```python
vec.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_movies[new_movies['title']==movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    
    for i in movies_list:
        print(new_movies.iloc[i[0]].title)
     
```
